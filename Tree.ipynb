# Trim a Binary Search Tree  

[669. Trim a Binary Search Tree](https://leetcode.com/problems/trim-a-binary-search-tree/)

**Solution Template:**

```python
class Solution(object):
    def func(self, root, L, R):
        
        # Case 1
        if not root: 
            return root
        
        # Case 2 - root deleted
        if root.val > R: 
            return self.trimBST(root.left, L, R)
        if root.val < L: 
            return self.trimBST(root.right, L, R)
        
        # Case 3 - root remained
        root.left = self.trimBST(root.left, L, R)
        root.right = self.trimBST(root.right, L, R)
        
        return root
```

# Level By Level Traversal 

- [107. Binary Tree Level Order Traversal II](https://leetcode.com/problems/binary-tree-level-order-traversal-ii/)
- [637. Average of Levels in Binary Tree](https://leetcode.com/problems/average-of-levels-in-binary-tree/)

This type of problem can be solved in two ways:
1. Using the "while current level not empty" technique
2. Recursive DFS 


**while loop Technique Template:**
```python
class Solution:
    def func(self, root):
        if not root:
            return []
        
        res = [] 
        current_level = [root] 
        
        while current_level:
            level_nodes = [] # Current Level Nodes
            next_level = [] # Next Level Nodes
            
            for node in current_level:
                level_nodes.append(node.val) 
                
                # Now we collect all the children nodes of the next level
                if node.left:
                    next_level.append(node.left)
                if node.right:
                    next_level.append(node.right) 
            
            res.append(level_nodes) # Update Result 
            current_level = next_level # Set Current Level to Next Level
        
        return res 
```

#  Minimum Distance Between BST Nodes

- [530. Minimum Absolute Difference in BST](https://leetcode.com/problems/minimum-absolute-difference-in-bst/) 
- [783. Minimum Distance Between BST Nodes](https://leetcode.com/problems/minimum-distance-between-bst-nodes/)

```python
class Solution:
    def getMinimumDifference(self, root): 
        
        # Step 1 - Initialize previous and res
        self.previous = float("inf")
        self.res = float("inf")
        
        # Step 2 - Call Auxiliary Function 
        self.aux(root)
        
        return self.res
    
    def aux(self, root):
        if not root:
            return 
        
        # Step 3 - While in aux, keep going left
        self.aux(root.left) 
        
        # Step 4 - Update res if newer value is smaller 
        self.res = min(self.res, abs(root.val - self.previous))
        
        # Step 5 - Set previous to be current
        self.previous = root.val
        
        # Step 6 - Go right
        self.aux(root.right)
```

# Sum of Left Leaves 

```python
class Solution(object):
    def sumOfLeftLeaves(self, root):
        if not root: 
            return 0
        
        # if root.left is a leave node:
        if root.left and not root.left.left and not root.left.right:
            # then recurse on root.right
            return root.left.val + self.sumOfLeftLeaves(root.right)
        
        # else, meaning neither left nor right is leaf
        return self.sumOfLeftLeaves(root.left) + self.sumOfLeftLeaves(root.right)
```

# Diameter of Binary Tree (REVIEW)

```python
class Solution(object):
    def diameterOfBinaryTree(self, root):
        self.res = 0
        self.aux(root)
        return self.res
    
    def aux(self, root):
        if not root: 
            return 0
        
        left = self.aux(root.left)
        right = self.aux(root.right)
        
        self.res = max(self.res, left + right)
        return 1 + max(left, right)
```

# Symmetric Tree 

```python
class Solution:
    def isSymmetric(self, root):
        if not root:
            return True
        
        return self.dfs(root.left, root.right)

    def dfs(self, l, r):
        if l and r:
            return l.val == r.val and self.dfs(l.left, r.right) and self.dfs(l.right, r.left)
        
        return l == r
```
